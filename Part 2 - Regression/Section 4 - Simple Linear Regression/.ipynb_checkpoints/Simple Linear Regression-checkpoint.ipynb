# Simple Linear Regression

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Salary_Data.csv')
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
print('X: {}'.format(X.shape))
print('y: {}'.format(y.shape))
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
print('X train: {} \n{}\n'.format(X_train.shape, X_train[:3,]))
print('y train: {} \n{}\n'.format(y_train.shape, y_train[:3,]))
print('X test: {} \n{}\n'.format(X_test.shape, X_test[:3,]))
print('y test: {} \n{}\n'.format(y_test.shape, y_test[:3,]))
```

```python
# Fitting Simple Linear Regression to the Training Set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
y_pred
```

```python
# What is the difference between prediction and test results
for i, x in enumerate(y_pred):
    diff = y_pred[i] - y_test[i]
    print('{} - {} = {}'.format(y_pred[i], y_test[i], diff))
```

```python
# Visualising the Training set results
plt.scatter(X_train, y_train, color='r')
train_pred = regressor.predict(X_train)
plt.plot(X_train, train_pred, color='b')
plt.title('Salary vs Experince Linear Regression (Training)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary in Dollars')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_train, y_train, color='r')
plt.scatter(X_test, y_test, color='g')
train_pred = regressor.predict(X_train)
plt.plot(X_train, train_pred, color='b')
plt.title('Salary vs Experince Linear Regression (Test)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary in Dollars')
plt.show()
```
